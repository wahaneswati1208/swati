```python
import pandas as pd
import numpy as np
```

```python
task =pd.read_csv("/content/task 2.csv")
task
```

#Data Integrity

```python
def validate_data(dataframe):
    dataframe['last_review'] = pd.to_datetime(dataframe['last_review'], errors='coerce')
    dataframe = dataframe.dropna(subset=['last_review'])
    return dataframe

df=validate_data(task)
```

#Missing Value

```python
task['reviews_per_month'].fillna(task['reviews_per_month'].mean(), inplace=True)
task.fillna(method='ffill', inplace=True)
task
```

#Duplicate Removal

```python
task.drop_duplicates(inplace=True)
task
```

#Standarization

```python
task['last_review'] = pd.to_datetime(task['last_review'], format='%Y-%m-%d')
task['room_type'] = task['room_type'].str.capitalize()
```

#OUTLIER

```python
from scipy import stats
z_scores = stats.zscore(task['price'])
abs_z_scores = np.abs(z_scores)
filtered_entries = (abs_z_scores < 3)  # Typically, a z-score > 3 is considered an outlier
df = task[filtered_entries]
z_scores

```
