#FEATURE SCALING

```python
features = df.columns.difference(['Class'])
target = 'Class'
```

```python
X=df[features]
y=df[target]
```

#TRAINING AND TESTING

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

#REAL TIME MONITORING

```python
import time
import random
def generate_transactions(max_iterations):
    iteration = 0
    while iteration < max_iterations:
        transaction_amount = random.uniform(1, 1000)
        transaction_time = time.strftime('%Y-%m-%d %H:%M:%S', time.localtime())
        yield {"amount": transaction_amount, "time": transaction_time}
        iteration += 1
def detect_fraud(transaction):
    if transaction["amount"] > 500:
        return True
    return False
max_iterations = 10
for transaction in generate_transactions(max_iterations):
    if detect_fraud(transaction):
        print(f"Potential fraud detected: {transaction}")
    else:
        print(f"Transaction processed: {transaction}")
    time.sleep(1)
```
